```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
#DATA CLEANING

## Missing values
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
```

```python
dataset[['Region']]=dataset[['Region']].astype(int)
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
## removing the null values
dataset=dataset.dropna().reset_index(drop=True)
```

```python
dataset.head()
```

```python
dataset.iloc[[122]]
```

```python
dataset=dataset.drop(122).reset_index(drop=True)
```

```python
dataset.iloc[[122]]
```

```python
dataset.columns
```

```python
# fix spaces in column names
dataset.columns=dataset.columns.str.strip()
dataset.columns
```

```python
##change the required columns as integer data type
dataset[['month','day','year','Temperature','RH','Ws']]=dataset[['month','day','year','Temperature','RH','Ws']].astype(int)
dataset.info()
```

```python
## changing the other columns to float data type

objects=[features for features in dataset.columns if dataset[features].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        dataset[i]=dataset[i].astype(float)
```

```python
dataset.info()
```

```python
#lets save the cleaned dataset
dataset.to_csv('Algerian_forest_fires_cleaned.csv',index=False)
```

```python
#drop day,m0nth,year
dataset_copy=dataset.drop(['day','month','year'],axis=1)
```

```python
#EDA

dataset_copy.head()
```

```python
# encoding of categories in classes

dataset_copy['Classes']=np.where(dataset_copy['Classes'].str.contains('not fire'),0,1)
```

```python
dataset_copy.head()
```

```python
#3categories in classes
dataset_copy['Classes'].value_counts()
```

```python
## plot density plot for all features
plt.style.use('seaborn-v0_8')  

dataset_copy.plot(kind='density', subplots=True, 
                  layout=(4,4), figsize=(20,15), sharex=False)

plt.show()

```

```python
## percentage for pie chart

percentage=dataset_copy['Classes'].value_counts(normalize=True)*100
```

```python
# plotting pir chart

classlabels=["Fire","Not fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title=("Pie Chart of Classes")
plt.show()
```
